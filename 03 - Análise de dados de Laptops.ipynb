### Renomear Colunas

```python
df = df.rename(columns={'Brand' : 'Marca', 
                        'Model' : 'Modelo',
                        'Screen' : 'Tela',
                        'Final Price':'Valor'})
df
```

### Limpeza dos dados

```python
''' Remoção de colunas não utilizadas
     -inplace True aplica no dataframe atual, False retorna uma cópia
     - axis: columns ou 1 aplica nas colunas, 0 ou index aplica no índice '''
    
df.drop(['Status', 'Touch'], inplace=True, axis='columns')
df.shape
```

```python
# dados nulos
df.isnull().sum()
```

- Não vou remover a coluna GPU, mesmo contendo muitos nulos pois quero saber quantos laptops possuem ou não uma GPU dedicada

### Padronização

```python
#Ajuste a GPU para listar as integradas ao invés de null
df['GPU'].fillna('Integrada', inplace=True)
```

### Gráfico das GPUs

```python
# Top 10 de GPUs
df.GPU.value_counts().head(10).plot(kind='bar')
```

```python
df_gpu=df['GPU'].value_counts()

plt.figure(figsize=(7, 5))
plt.pie(df_gpu,labels=df_gpu.index,autopct='%1.f%%')
plt.title('Distribuição de GPUs')
plt.show()
```

```python
# Agrupar todos os itens menores de 10 em uma categoria OUTROS

df_gpu = df['GPU'].value_counts()

regra_agrupamento = df_gpu[df_gpu < 10].index

# Usa uma lambda pra pegar todos os itens menores do que a regra e agrupar
df['GPU_agrupada'] = df['GPU'].apply(lambda x: 'Outros' if x in regra_agrupamento else x)

# Conta o número de ocorrências após o agrupamento
df_gpu_agrupada = df['GPU_agrupada'].value_counts()

# Configuração do gráfico de pizza
plt.figure(figsize=(7, 5))
plt.pie(df_gpu_agrupada, labels=df_gpu_agrupada.index, autopct='%1.f%%')
plt.title('Distribuição de GPUs')
plt.show()
```

### Heatmap

```python
# Agregação com a média de valor entre duas colunas do dataframe
df_agregado = df.groupby(['Marca', 'GPU'])['Valor'].mean().reset_index().nlargest(100, 'Valor')

df_pivot = df_agregado.pivot(index='Marca', columns='GPU', values='Valor')

df_pivot.fillna(0, inplace=True)

df_pivot.sample(5)
```

```python
f, ax = plt.subplots(figsize=(12, 5))


sns.heatmap(df_pivot, annot=False, linewidths=.5, ax=ax)
plt.title('Heatmap de Valor por GPU e Marca')
plt.show()
```
